```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# True values
y_true = torch.tensor([1.0, 2.0, 3.0])

# Different predicted values with increasing error
y_preds = [
    torch.tensor([1.0, 2.0, 3.0]),  # perfect
    torch.tensor([1.5, 2.5, 3.5]),  # small error
    torch.tensor([2.0, 3.0, 4.0]),  # medium error
    torch.tensor([3.0, 4.0, 5.0])   # large error
]

# Calculate MSE losses
mse_losses = [((y_true - y_pred) ** 2).mean().item() for y_pred in y_preds]

# X values for plotting
x = np.arange(len(y_true))

# Plot
fig, axs = plt.subplots(1, len(y_preds), figsize=(16, 4))

for i, (y_pred, loss) in enumerate(zip(y_preds, mse_losses)):
    axs[i].plot(x, y_true, label='True', marker='o')
    axs[i].plot(x, y_pred, label='Predicted', marker='x')
    axs[i].set_title(f'MSE Loss = {loss:.2f}')
    axs[i].set_ylim(0, 6)
    axs[i].legend()
    axs[i].grid(True)

fig.suptitle("Effect of Prediction Error on MSE Loss", fontsize=16)
plt.tight_layout()
plt.show()
```
