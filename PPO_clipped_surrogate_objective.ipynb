```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
epsilon = 0.2
advantage = 1.0

# Range of probability ratios
r = np.linspace(0.5, 1.5, 500)

# Unclipped objective
obj_unclipped = r * advantage

# Clipped ratio
r_clipped = np.clip(r, 1 - epsilon, 1 + epsilon)
obj_clipped = np.minimum(obj_unclipped, r_clipped * advantage)

# Plot
plt.figure(figsize=(8,5))
plt.plot(r, obj_unclipped, label='Unclipped objective $r \\hat{A}$', color='blue')
plt.plot(r, obj_clipped, label='Clipped surrogate objective', color='orange')
plt.axvline(1 - epsilon, color='red', linestyle='--', label='$1 - \\epsilon$ clip threshold')
plt.axvline(1 + epsilon, color='green', linestyle='--', label='$1 + \\epsilon$ clip threshold')
plt.axhline(advantage, color='gray', linestyle=':', label='Advantage $\\hat{A}$')

plt.title('PPO Clipped Surrogate Objective vs Probability Ratio $r$')
plt.xlabel('Probability Ratio $r = \\frac{\\pi_\\theta}{\\pi_{\\theta_{old}}}$')
plt.ylabel('Objective Value')
plt.legend()
plt.grid(True)
plt.show()
```
